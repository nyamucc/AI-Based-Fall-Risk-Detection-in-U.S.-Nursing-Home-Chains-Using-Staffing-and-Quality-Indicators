#### Importing Libraries

```python
import pandas as pd
```

```python
df=pd.read_csv('RISKS.csv')
```

```python
df.head(10)
```

##### Removing first row

```python
df = df[df["Chain"] != "National"]
```

#### Removing unwanted columns

```python
df.drop(columns=
        ['Chain', 'Chain ID', 'Number of states and territories with operations',
    'Number of Special Focus Facilities (SFF)', 'Number of SFF candidates',
    'Number of facilities with an abuse icon', 'Percentage of facilities with an abuse icon',
    'Percent of facilities classified as for-profit', 'Percent of facilities classified as non-profit',
    'Percent of facilities classified as government-owned', 'Average total weekend nurse hours per resident day',
    'Average Registered Nurse turnover percentage', 'Average number of administrators who have left the nursing home',
    'Average number of fines', 'Total amount of fines in dollars', 'Average amount of fines in dollars',
    'Average number of payment denials', 'Average percentage of short-stay residents who were re-hospitalized after a nursing home admission',
    'Average percentage of short-stay residents who have had an outpatient emergency department visit',
    'Average percentage of short-stay residents who newly received an antipsychotic medication',
    'Average percentage of short-stay residents with pressure ulcers or pressure injuries that are new or worsened',
    'Average percentage of short-stay residents who are at or above an expected ability to care for themselves and move around at discharge ',
    'Average percentage of short-stay residents who were assessed and appropriately given the seasonal influenza vaccine',
    'Average percentage of short-stay residents who were assessed and appropriately given the  pneumococcal vaccine',
    'Average number of outpatient emergency department visits per 1,000 long-stay resident days',
    'Average percentage of long-stay residents who received an antipsychotic medication',
    'Average percentage of long-stay residents who were assessed and appropriately given the seasonal influenza vaccine',
    'Average percentage of long-stay residents who were assessed and appropriately given the  pneumococcal vaccine',
    'Average percentage of long-stay residents who used antianxiety or hypnotic medication',
    'Average rate of potentially preventable hospital readmissions 30 days after discharge from a SNF'], inplace=True)
```

```python
df.columns.tolist()
```

#### Renaming Columns

```python
df.rename(columns={
    'Number of facilities': 'num_facilities',
    'Average overall 5-star rating': 'overall_rating',
    'Average health inspection rating': 'health_rating',
    'Average staffing rating': 'staffing_rating',
    'Average quality rating': 'quality_rating',
    'Average total nurse hours per resident day': 'nurse_hours_total',
    'Average total Registered Nurse hours per resident day': 'rn_hours_total',
    'Average total nursing staff turnover percentage': 'staff_turnover_pct',
    'Total number of fines': 'total_fines',
    'Total number of payment denials': 'total_payment_denials',
    'Average number of hospitalizations per 1,000 long-stay resident days': 'hosp_per_1000_residents',
    'Average percentage of long-stay residents experiencing one or more falls with major injury': 'falls_with_injury',
    'Average percentage of long-stay residents with pressure ulcers': 'pressure_ulcers',
    'Average percentage of long-stay residents with a urinary tract infection': 'uti_pct',
    'Average percentage of long-stay residents who have or had a catheter inserted and left in their bladder': 'catheter_pct',
    'Average percentage of long-stay residents whose ability to move independently worsened': 'mobility_decline_pct',
    'Average percentage of long-stay residents whose need for help with activities of daily living has increased': 'adl_increase_pct',
    'Average percentage of long-stay residents who were physically restrained': 'restrained_pct',
    'Average percentage of long-stay residents with new or worsened bowel or bladder incontinence': 'incontinence_pct',
    'Average percentage of long-stay residents who lose too much weight': 'weight_loss_pct',
    'Average percentage of long-stay residents who have symptoms of depression': 'depression_pct'
}, inplace=True)
```
